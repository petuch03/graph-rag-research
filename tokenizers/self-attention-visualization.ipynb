```python
# Load model and retrieve attention weights

from bertviz import head_view, model_view
from transformers import GemmaModel, GemmaConfig, GemmaTokenizer, GemmaForCausalLM

# configuration = GemmaConfig()
model_version = 'google/gemma-2b-it'

model = GemmaForCausalLM.from_pretrained(model_version, output_attentions=True)



# model = GemmaModel.from_pretrained(model_version, output_attentions=True)
tokenizer = GemmaTokenizer.from_pretrained(model_version)
```

```python
input_text = 'Bob sent Alice a message about apples.'
inputs = tokenizer.encode(input_text, return_tensors='pt')  # Tokenize input text
tokens = tokenizer.convert_ids_to_tokens(inputs[0])
print(inputs)
print(tokens)
```

```python
outputs = model(inputs)  # Run model
attention = outputs[-1]
model_view(attention, tokens)  # Display model view
```

```python
to_tokenize = ['Bob', '▁se', 'nt', '▁Alice', '▁a', '▁mes', 'sage', '▁ab', 'out', '▁app', 'les','.']
print("\"" + ''.join(to_tokenize) + "\"")
sub_word_tokens = tokenizer.encode(to_tokenize, return_tensors='pt')
tokens = tokenizer.convert_ids_to_tokens(sub_word_tokens[0])
print(sub_word_tokens)
```

```python
outputs = model(sub_word_tokens)  # Run model
attention = outputs[-1]
model_view(attention, tokens)  # Display model view
```
